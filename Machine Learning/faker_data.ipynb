```python
from faker import Faker
import pandas as pd
```

```python
df = pd.read_csv('dataset/data_cleaned.csv')
fake = Faker('id_ID')
```

```python
new_df = df[["job_title", "job_description"]]
new_df = new_df.rename(columns={"job_description": "description", "job_title": "jobTitle"})
new_df
```

```python
company_names = []
for _ in range(24824):
    company_name = fake.company()
    company_names.append(company_name)

faker_df = pd.DataFrame({'companyName': company_names})
faker_df
```

```python
addresses = []
for _ in range(24824):
    address = fake.address()
    addresses.append(address.split("\n")[0])

faker_df['address'] = addresses
faker_df
```

```python
phoneNumbers = []
for _ in range(24824):
    phoneNumber = fake.phone_number()
    phoneNumbers.append(phoneNumber)

faker_df['phoneNumber'] = phoneNumbers
faker_df
```

```python
def generate_email(company_name):
    parsed_name = company_name.replace(" ", "").lower()
    email = parsed_name + "@email.com"
    return email

faker_df['email'] = faker_df['companyName'].apply(generate_email)
faker_df
```

```python
def generate_website(company_name):
    parsed_name = company_name.replace(" ", "-").lower()
    website = "www." + parsed_name + ".com"
    return website

faker_df['website'] = faker_df['companyName'].apply(generate_website)
faker_df
```

```python
def generate_linkedin(company_name):
    parsed_name = company_name.replace(" ", "-").lower()
    linkedin = "www.linkedin.com/in/" + parsed_name + "/"
    return linkedin

faker_df['linkedin'] = faker_df['companyName'].apply(generate_linkedin)
faker_df
```

```python
userId = []
for _ in range(24824):
    user_id = "e082faa8-ef7f-4f07-910d-b87d6a2c6b92"
    userId.append(user_id)

user_id_df = pd.DataFrame({'userId': userId})
user_id
```

```python
drop_df = df.drop(['job_title', 'job_description'], axis=1)

concatenated_df = pd.concat([user_id_df, new_df, faker_df, drop_df], axis=1)
concatenated_df
```

```python
concatenated_df.to_csv("dataset/data_cleaned_v2.csv", index=False)
```
